```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
flower = pd.DataFrame(iris.data, columns=iris.feature_names)
flower.head()
```

```python
flower['target']=iris.target
flower[0:56]
```

```python
flower0 = flower[:50]
flower1 = flower[50:100]
flower2 = flower[100:]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(flower0['sepal length (cm)'],flower0['sepal width (cm)'], color = 'green', marker='o')
plt.scatter(flower1['sepal length (cm)'],flower1['sepal width (cm)'], color = 'red', marker='o')
plt.scatter(flower2['sepal length (cm)'],flower2['sepal width (cm)'], color = 'blue', marker='o')
```

```python
plt.scatter(flower0['petal length (cm)'],flower0['petal width (cm)'], color = 'green', marker='o')
plt.scatter(flower1['petal length (cm)'],flower1['petal width (cm)'], color = 'red', marker='o')
plt.scatter(flower2['petal length (cm)'],flower2['petal width (cm)'], color = 'blue', marker='o')
```

```python
# Creating train dataset

X = flower.drop('target',axis = 'columns')
y = flower.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 10)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```
